```python
# @title ‚öúÔ∏è C√†i ƒë·∫∑t v√† T·∫£i Model
import os
import sys
import requests
from tqdm.notebook import tqdm

# =================================================================
# B∆Ø·ªöC 1: T·∫¢I D·ª∞ √ÅN "ANIGA3" T·ª™ GITHUB
# =================================================================
project_dir = "/content/Aniga3"
repo_url = "https://github.com/jofix2004/Aniga3.git"

if os.path.exists(project_dir):
    print("‚úÖ Th∆∞ m·ª•c d·ª± √°n Aniga3 ƒë√£ t·ªìn t·∫°i. B·ªè qua b∆∞·ªõc t·∫£i l·∫°i.")
else:
    print("B∆Ø·ªöC 1: ƒêang t·∫£i d·ª± √°n Aniga3...")
    !git clone {repo_url} {project_dir}
    print("‚úÖ T·∫£i d·ª± √°n ho√†n t·∫•t!")

os.chdir(project_dir)
if project_dir not in sys.path:
    sys.path.append(project_dir)

# =================================================================
# B∆Ø·ªöC 2: C√ÄI ƒê·∫∂T C√ÅC TH∆Ø VI·ªÜN PYTHON
# =================================================================
print(f"\nB∆Ø·ªöC 2: ƒêang c√†i ƒë·∫∑t c√°c th∆∞ vi·ªán...")
!pip install -q torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu118
!pip install -q -r requirements.txt
print("‚úÖ C√†i ƒë·∫∑t th∆∞ vi·ªán ho√†n t·∫•t!")

# =================================================================
# B∆Ø·ªöC 3: T·∫¢I C√ÅC MODEL ƒê√É HU·∫§N LUY·ªÜN
# =================================================================
models_to_download = {
    "model1.pt": "https://huggingface.co/TranLinh2004/AniGa-CleMove/resolve/main/v8-m3.pt?download=true",
    "model2.pt": "https://huggingface.co/TranLinh2004/AniGa-CleMove/resolve/main/v9-c2_fp16.pt?download=true",
    "model3.pt": "https://huggingface.co/TranLinh2004/AniGa-CleMove/resolve/main/v11-m3.pt?download=true"
}

def download_file(url, destination):
    """T·∫£i file v·ªõi thanh ti·∫øn tr√¨nh."""
    if os.path.exists(destination):
        print(f"‚úÖ File {os.path.basename(destination)} ƒë√£ t·ªìn t·∫°i. B·ªè qua.")
        return True
    try:
        with requests.get(url, stream=True, timeout=60) as r:
            r.raise_for_status()
            total_size = int(r.headers.get('content-length', 0))
            with open(destination, 'wb') as f, tqdm(
                total=total_size, unit='iB', unit_scale=True,
                desc=f"ƒêang t·∫£i {os.path.basename(destination)}"
            ) as bar:
                for chunk in r.iter_content(chunk_size=8192):
                    f.write(chunk)
                    bar.update(len(chunk))
        print(f"‚úÖ T·∫£i {os.path.basename(destination)} th√†nh c√¥ng!")
        return True
    except requests.exceptions.RequestException as e:
        print(f"‚ùå L·ªñI khi t·∫£i {url}: {e}")
        return False

print("\nB∆Ø·ªöC 3: --- B·∫ÆT ƒê·∫¶U QU√Å TR√åNH T·∫¢I MODEL ---")
all_successful = True
for filename, url in models_to_download.items():
    destination_path = os.path.join(project_dir, filename)
    if not download_file(url, destination_path):
        all_successful = False

if all_successful:
    print("\n--- ‚úÖ M√îI TR∆Ø·ªúNG ƒê√É S·∫¥N S√ÄNG ---")
    print("B·∫°n c√≥ th·ªÉ ch·∫°y √¥ ti·∫øp theo ƒë·ªÉ kh·ªüi ƒë·ªông ·ª©ng d·ª•ng.")
else:
    print("\n--- ‚ùå C√ì L·ªñI X·∫¢Y RA KHI T·∫¢I MODEL ---")
    print("Vui l√≤ng ki·ªÉm tra l·∫°i l·ªói v√† th·ª≠ ch·∫°y l·∫°i √¥ n√†y.")
```
